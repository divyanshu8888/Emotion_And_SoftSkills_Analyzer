```python
import cv2
from transformers import pipeline
from deepface import DeepFace
import speech_recognition as sr
import threading
import numpy as np

# Initialize NLP pipeline for soft skills (sentiment analysis)
soft_skill_pipeline = pipeline("sentiment-analysis")

# Initialize speech recognizer
recognizer = sr.Recognizer()

# Global variables
transcribed_text = ""
soft_skills_results = []
emotion_results = []
recording = False
frame_counter = 0
emotion_analysis_interval = 5  # Perform emotion analysis every 5 frames (to reduce load)
current_emotion = ""  # Store the current emotion to display for longer

# Load pre-trained face detector (Haar Cascade)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Function to start video capture and perform emotion analysis
def capture_video():
    global recording, emotion_results, frame_counter, current_emotion

    # Open webcam with higher resolution
    cap = cv2.VideoCapture(0)
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)  # Set video frame width
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)  # Set video frame height
    if not cap.isOpened():
        print("Error: Could not open video source.")
        return

    print("Recording started... Press 'q' to Stop. Press 'c' to capture screenshot.")
    
    while recording:
        ret, frame = cap.read()
        if not ret:
            print("Failed to capture frame. Exiting...")
            break

        # Perform face detection
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.1, 4)

        # Perform emotion analysis on each frame every 'emotion_analysis_interval' frames
        if frame_counter % emotion_analysis_interval == 0:
            try:
                # Perform emotion analysis on the frame
                emotions = DeepFace.analyze(frame, actions=['emotion'], enforce_detection=False)
                dominant_emotion = emotions[0]['dominant_emotion']
                emotion_results.append(dominant_emotion)
                current_emotion = dominant_emotion  # Update the current emotion
                
                # Check if looking straight into the camera (confident)
                if len(faces) > 0:
                    (x, y, w, h) = faces[0]
                    face_center = (x + w / 2, y + h / 2)
                    frame_center = (frame.shape[1] / 2, frame.shape[0] / 2)
                    # Simple logic: if face is near the center of the frame
                    if abs(face_center[0] - frame_center[0]) < 50 and abs(face_center[1] - frame_center[1]) < 50:
                        cv2.putText(frame, "Emotion: Confident", (10, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2, cv2.LINE_AA)
                    else:
                        cv2.putText(frame, f"Emotion: {dominant_emotion}", (10, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2, cv2.LINE_AA)
                else:
                    cv2.putText(frame, f"Emotion: {dominant_emotion}", (10, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2, cv2.LINE_AA)
            except Exception as e:
                print("Emotion analysis error:", e)

        # Display the bounding box around the face
        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

        # Display the transcribed speech on the frame
        cv2.putText(frame, f"Speech: {transcribed_text[-50:]}", (10, 100), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
        
        # Display the emotion text for a longer duration
        if current_emotion:
            cv2.putText(frame, f"Emotion: {current_emotion}", (10, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2, cv2.LINE_AA)
        
        cv2.imshow("Video Capture & Emotion Analysis", frame)

        # Check for key press actions
        key = cv2.waitKey(1) & 0xFF
        if key == ord('q'):  # Stop recording when 'q' is pressed
            print("Stopping recording...")
            recording = False
            break
        elif key == ord('c'):  # Capture screenshot when 'c' is pressed
            screenshot_filename = f"emotion_screenshot_{current_emotion}.png"
            cv2.imwrite(screenshot_filename, frame)
            print(f"Screenshot saved as {screenshot_filename}")

        frame_counter += 1

    # Release resources after recording stops
    cap.release()
    cv2.destroyAllWindows()

# Function to transcribe audio and analyze soft skills in real-time
def transcribe_audio():
    global transcribed_text, soft_skills_results, recording

    with sr.Microphone() as source:
        recognizer.adjust_for_ambient_noise(source)
        print("Listening for speech...")

        while recording:
            try:
                audio = recognizer.listen(source, phrase_time_limit=5)
                text = recognizer.recognize_google(audio)
                transcribed_text += " " + text
                print(f"Transcribed Speech: {text}")

                # Perform sentiment analysis for soft skills
                soft_skill_result = soft_skill_pipeline(text)[0]
                soft_skills_results.append(soft_skill_result)
                print(f"Soft Skills Analysis: {soft_skill_result}")
            except sr.UnknownValueError:
                print("Could not understand audio.")
            except sr.RequestError as e:
                print(f"Could not request results from Google Speech Recognition service; {e}")

# Final Soft Skills Result Decision
def get_final_soft_skills_result():
    negative_count = sum(1 for result in soft_skills_results if result['label'] == 'NEGATIVE')
    positive_count = sum(1 for result in soft_skills_results if result['label'] == 'POSITIVE')

    if negative_count > positive_count:
        return {'label': 'NEGATIVE', 'score': 1.0}
    else:
        return {'label': 'POSITIVE', 'score': 1.0}

# Main function to initiate recording and analysis
def start_recording():
    global recording
    recording = True

    # Start video and audio capture simultaneously in separate threads
    video_thread = threading.Thread(target=capture_video)
    audio_thread = threading.Thread(target=transcribe_audio)
    video_thread.start()
    audio_thread.start()

    # Wait for both threads to finish
    video_thread.join()
    audio_thread.join()

    # Get final soft skills result after processing all inputs
    final_soft_skills_result = get_final_soft_skills_result()

    # Display final results
    print("\nFinal Transcribed Speech:", transcribed_text)
    print("Final Emotion Results:", emotion_results)
    print("Final Soft Skills Result:", final_soft_skills_result)

# Run the recording function
start_recording()
```
